```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
city_wise_sales_mean = round(data.groupby('City')['Sales'].mean(),2)
city_wise_sales_mean
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=city_wise_sales_mean.values, y=city_wise_sales_mean.index, palette=sns.color_palette())
plt.title("City wise sales mean")
plt.xlabel("Sales mean")
plt.ylabel("City")
plt.tight_layout()
plt.show()
```

```python
region_wise_discount=round(data.groupby('Region')['Discount'].mean(),2)
region_wise_discount
```

```python
plt.figure(figsize=(5,5))
sns.barplot(x=region_wise_discount.index, y=region_wise_discount.values, palette=sns.color_palette())
plt.title("Region wise Discount Mean")
plt.ylabel("Discount mean")
plt.xlabel("Region")
plt.tight_layout()
plt.show()
```

```python
data.head()
```

```python
product_wise_sales_mean=round(data.groupby('Category')['Sales'].mean(),2)
product_wise_sales_mean
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=product_wise_sales_mean.index, y=product_wise_sales_mean.values, palette=sns.color_palette())
plt.title("Product wise sales Mean")
plt.ylabel("Sales mean")
plt.xlabel("Product")
plt.tight_layout()
plt.show()
```

```python
product_wise_discount_mean=round(data.groupby('Category')['Discount'].mean(),2)
product_wise_discount_mean
```

```python
plt.figure(figsize=(10,6))
plt.pie(product_wise_discount_mean.values, labels = product_wise_discount_mean.index, autopct='%1.1f%%', startangle=90)
plt.title("Product wise Discount Mean")
plt.show()
```

```python
print(data['Order Date'].head())
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'], infer_datetime_format=True, errors='coerce')
data['Order Date'] = pd.to_datetime(data['Order Date'], format='%m/%d/%Y', errors='coerce')
data['Order Year'] = data['Order Date'].dt.year
print(data.head())
```

```python
data.head()
```

```python
Profit_year=round(data.groupby('Order Year')['Profit'].sum(),2)
Profit_year
```

```python
plt.figure(figsize=(10,5))
plt.pie(Profit_year.values, labels = Profit_year.index, autopct='%1.1f%%', startangle=90)
plt.title("Profit Year")
plt.show()
```
